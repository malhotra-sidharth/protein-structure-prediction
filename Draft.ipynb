```python
import pandas as pd
import numpy as np
from biopandas.pdb import PandasPdb as ppd
```

```python
data = ppd().read_pdb('./data/6ewu.pdb')
```

```python
data.df['OTHERS'][data.df['OTHERS']['record_name'] == 'HELIX']
```
